Załadowanie potrzebnych bibliotek:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

Załadowanie danych o irysach:

```python
# load the famous iris data
irisRaw = load_iris()

# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

Zad 4. - metoda groupby(). Metoda ta pozwala na pogrupowanie danych względem jednej z wartości.

W ramach ćwiczenia pogrupowano rekordy względem wartości w kolumnie 'sepal length (cm)' i obliczono średnią dla pozostałych parametrów.

```python
iris.groupby(['sepal length (cm)']).mean()
```

Zad. 6 - metoda dropna(). Metoda ta usuwa z DataFrame'a wiersze, w których znajdują się wartości NaN.

Dane nie zawierają braków, co sprawdzono po liczbie wierszy przed i po użyciu metody dropna()

```python
iris_drop_na = iris.dropna()
```

```python
print(f"Liczba wierszy przed zastosowaniem dropna():  {iris.size}")
print(f"Liczba wierszy po zastosowaniu dropna():  {iris_drop_na.size}")
```

Zad 7. - sortowanie względem kolumny 'sepal width'.

```python
iris.sort_values(['sepal width (cm)'],ascending=True)
```

Zad. 8. - wyznaczanie maximów i minimów z kolumny 'petal length (cm)'

```python
min_petal, idxmin = iris['petal length (cm)'].min(), iris['petal length (cm)'].idxmin()
print(f"Petal minimum: {min_petal},\nmin_id: {idxmin}")
```

```python
max_petal, idxmax = iris['petal length (cm)'].max(), iris['petal length (cm)'].idxmax()
print(f"Petal maximum: {max_petal},\nmax_id: {idxmax}")
```

Zad. 9. - obliczenie ochyleń standardowych dla poszczególnych kolumn

```python
iris.std()
```

Zad. 10. - Wyodrębnienie wierszy dla których długość kielicha kwiatów (sepal length) jest większa od średniej długości.

```python
iris[iris['sepal length (cm)'] > iris['sepal length (cm)'].mean()]
```

Zad 11. - histogramy danych

```python

for param in iris.columns:
    plt.figure()
    iris[param].hist()
    plt.title(f'Histogram of: {param}')
    plt.xlabel(f'{param}')
    plt.ylabel('Number of records')
    plt.show()
```
